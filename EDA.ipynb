## exploratory data analysis

```python
# importando bibliotecas

import pandas as pd
import numpy as np
```

```python
# lendo base de dados

df = pd.read_csv("sales.csv", low_memory=False)
```

```python
# visualizando base de dados

pd.set_option("display.max_columns", None)
df.head()
```

```python
df.bi_st.value_counts()
```

```python
df[["status", "bi_st"]]
```

```python
df.groupby(["bi_st", "status"])["total"].sum()
```

```python
df["Name Prefix"].unique()
```

```python
# identificando numero de compra de clientes

df.query("full_name == 'Harper, Alvaro'")['order_id'].count()
```

```python
df.query('full_name == "Harper, Alvaro"')
```

```python
df.query("full_name == 'Harper, Alvaro' and status != 'canceled'")
```

```python
# categorias que cada genero compra mais

df.groupby(['Gender', 'category'])['total'].sum().reset_index().sort_values(by=['Gender', 'total'], ascending=False)                                                                   
```

```python
# identificando qual genero compra mais por categoria

df.groupby(['category', 'Gender'])['Gender'].count()
```

```python
# identificar o período de tempo dessa base de dados

df.year.unique()
```

```python
df.groupby(['year', 'month'])['total'].sum()
```

```python
df.order_date.nunique()
```

```python
df.groupby('status')['order_id'].count().reset_index().sort_values(by='order_id', ascending=False)
```

```python
df.query('status == "canceled"')['payment_method'].value_counts()
```

```python
df.query('status == "complete"')['payment_method'].value_counts()
```

```python
df.query('status == "canceled"')['category'].value_counts()
```

```python
# top 10 de produtos mais cancelados 


df.query('status == "canceled"').groupby(['category','sku'])['order_id'].count().reset_index().sort_values(by='order_id', ascending=False).head(10)
```

```python
#Estados que mais arrecadam com vendas

df.groupby('State')['total'].sum().reset_index().sort_values(by='total', ascending=False)
```

```python
# No TX, qual as cidades que mais vendem

df.query('State == "TX"').groupby('City')['total'].sum().reset_index().sort_values(by='total', ascending=False)
```

```python
# Na cidade que mais vende, qual a categoria mais vendidos

df.query('City == "Houston"').groupby('category')['total'].sum().reset_index().sort_values(by='total', ascending=False)
```

```python
# media das % de desconto por categoria

df.groupby('category')['Discount_Percent'].mean().reset_index().sort_values(by='Discount_Percent', ascending=False)
```
